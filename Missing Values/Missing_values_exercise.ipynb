```python
# STEP 1: Upload the dataset
# Prompt the user to manually select and upload the `sensor_log.csv` file
from google.colab import files
uploaded = files.upload()

# STEP 2: Load the dataset into a DataFrame
# Read the uploaded CSV file into a pandas DataFrame
import pandas as pd
df = pd.read_csv("sensor_log.csv")
print("Original Dataset:")
display(df)

# STEP 3: Assess missing values in each column
# Display the total number of missing values per column
print("\nMissing Values per Column:")
print(df.isnull().sum())

# STEP 4: Generate four different imputation techniques
# 1. Forward Fill (propagates the last valid observation forward)
df_forward_fill = df.fillna(method='ffill')

# 2. Backward Fill (propagates the next valid observation backward)
df_back_fill = df.fillna(method='bfill')

# 3. Interpolation (estimates values based on neighbouring data points)
df_interpolate = df.interpolate()

# 4. Mean Imputation (fills missing values with the column mean)
df_mean_fill = df.fillna(df.mean(numeric_only=True))

# STEP 5: Display the results of each imputation method
print("\n--- Forward Fill Result ---")
display(df_forward_fill)

print("\n--- Backward Fill Result ---")
display(df_back_fill)

print("\n--- Interpolation Result ---")
display(df_interpolate)

print("\n--- Mean Fill Result ---")
display(df_mean_fill)

# STEP 6: Export each cleaned dataset as a separate CSV file
# These files may be submitted or used for further analysis
df_forward_fill.to_csv("forward_fill.csv", index=False)
df_back_fill.to_csv("backward_fill.csv", index=False)
df_interpolate.to_csv("interpolated.csv", index=False)
df_mean_fill.to_csv("mean_fill.csv", index=False)

print("\nAll cleaned datasets have been successfully saved as CSV files.")
```

## Exercise 3  Forward Fill vs Interpolation (Time-Series Style)

**Task:**
1. Create a time-series style version of the dataset called `df_ts`.  
2. Create three dataframes from `df_ts`:  
   - one using **forward fill**,  
   - one using **interpolation**.  
3. For a small time range, compare the values side by side.  
4. Decide which method is most reasonable for **sensor data** and why.


**Discussion/Reflection:**  
- **Forward fill** repeats the last known sensor reading. This can be okay if values change slowly, but it may be unrealistic over long gaps.  
- **Interpolation** estimates missing values between two known readings and follows the trend of the data.  

For continuous sensor data such as temperature, humidity, and voltage, **interpolation is usually more reasonable**, because it preserves smooth changes over time instead of creating flat stretches.
